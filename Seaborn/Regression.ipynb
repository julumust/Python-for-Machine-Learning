### **Regression**

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### **Method -** `lmplot()`

`lmplot()` allows you to display linear models and split these plots up based off of features.

`hue='input'` separates cateogrical data.

`markers=['i','j']` adds a marker to each category found within the plot, in this case their are two, male and female.


```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

If visibility is an issue, one can add a `scatter_kws=` parameter. Seaborn calls matplotlib with `lmplot()`. To directly affect matplotlibs functionality, a dictionary can be passed to change the marker size.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

### **Using a Grid**

We can add more variable separation through columns and rows with the use of a grid. This can be done with `col` and `row` arguements.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

### **Aspect and Height**

Seaborn figures can have their height and aspect ratio adjusted with `aspect` and `height` formerly `size` but is since deprecated.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,height=8)
```
