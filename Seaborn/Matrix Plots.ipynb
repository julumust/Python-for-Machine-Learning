## **Matrix Plots**

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

### **Method -** `heatmap()`

To use `heatmap()` one must first ensure their data is in a matrix form.

**Note:** ensure to pass `numeric_only=True` to prevent str data from being passed into the function call. 

```python
tc = tips.corr(numeric_only=True)
sns.heatmap(tc,annot=True)
```

### **Heatmap for the flights dataset**

1. Convert the data into a matrix format as seen with `pivot_table()` below.
2. Plot your heatmap!

`linecolor='input'` changes map colour

`linewidth=int` creates a specified separation between the cells.

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(fp,cmap='magma',linecolor='white',linewidth=1)
```

### **Method -** `clustermap()`

`clustermap()` uses hierarchical clustering to produce a clustered version of the heatmap.

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

The output above no longer groups data chronologically and instead groups data according to value similiarity, in this case, passenger count.
