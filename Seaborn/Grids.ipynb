## **Grids**

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

### **Method -** `pairplot()`

This function plots pairwise relationships in a dataset. It is can be considered the simpler version of `PairGrid()`

`pairplot()` is a simpler, faster function for standard use cases, if one is looking for more flexiblity, `PairGrid()` offers full control over each subplot.

```python
sns.pairplot(iris)
```

### **Method -** `PairGrid()`

`PairGrid()` is a class that offers the user full contorl over each subplot.

```python
# Just the Grid
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)

# Here we map to the grid

g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

### **Method -** `FaceGrid()`

Multi-plot grid for plotting conditions relationships.

```python
tips = sns.load_dataset('tips')
```

```python
# Just the grid
sns.FacetGrid(data=tips,col='time',row='smoker')
```

### **Graph Customisation**

To separate categorical data, one can pass `hue='sex'` into the FacetGrid method.

```python
i = sns.FacetGrid(data=tips,col='time',row='smoker',hue='sex')

# Mapping data
i.map(plt.scatter,'total_bill','tip')
```

###**Method -** `JointGrid()'

`JointGrid` is the general version of `jointplot()` type grids.

```python
# Generate the figure
i = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
i = sns.JointGrid(x="total_bill", y="tip", data=tips)
i = i.plot(sns.regplot, sns.histplot)
```
