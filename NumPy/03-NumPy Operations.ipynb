### **NumPy Operations**

```python
import numpy as np
```

Using **mathemtical operators** to perform array arithmetic.

```python
arr = np.arange(0,11)
```

```python
arr + arr
```

```python
arr - arr
```

```python
arr * arr
```

```python
arr + 100
```

```python
arr * 100
```

#### Python will prompt you with a warning when dividing by zero!

#### It will replace errors with `nan`

```python
arr/arr
```

```python
# Also warning, but not an error instead infinity
1/arr
```

```python
arr ** 2
```

### **Universal Array Functions**

```python
# Taking square roots
np.sqrt(arr)
```

```python
# Calculating exponential (e^)
np.exp(arr)
```

The same as calling `.max()`

```python
np.max(arr)
```

### **Trignometric Operations**

```python
np.sin(arr)
```

```python
np.log(arr)
```
