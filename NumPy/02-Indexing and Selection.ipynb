## **Numpy Indexing and Selection**

```python
import numpy as np
```

Creating a sample array:

```python
arr = np.arange(0,11)
```

Display the array:

```python
arr
```

### **Bracket Indexing and Selection**

Get a value at an index by calling `arr[]`

```python
arr[8]
```

Get values in a range by calling `arr[:]`

```python
arr[1:5]
```

```python
arr[0:5]
```

```python
arr[0:6]
```

```python
arr[5:]
```

```python
arr[0:6]
```

### **Broadcasting**
NumPy lists differ from python lists because of their ability to **broadcast**.

```python
# Setting a value with index range --> Broadcasting!
arr[0:5] = 100
arr
```

```python
arr[0:5]
```

```python
arr = np.arange(0,11)
arr
```

```python
# Slicing Arrays
slice_of_arr = arr[0:6]
slice_of_arr
```

```python
# Change Slice
slice_of_arr[:] = 99
slice_of_arr
```

```python
arr
```

The original array remains unchanged. This avoids memory problems.

To get a copy of an array, you need to be explicit by calling on `.copy()`

```python
arr_copy = arr.copy()
arr
```

```python
arr_copy[:] = 100
arr_copy
```

```python
arr
```

### **Indexing a 2D array (matrices)**

```python
arr_2d = np.array([[5,10,15],[20,25,30],[35,40,45]])
arr_2d
```

To get an individual element use: `arr_2d[index][index]` or `arr_2d[index,index]`

Comma notation is recommended for **clarity**.

```python
# Getting individual element value
arr_2d[0][0]
```

```python
# Getting individual element value
arr_2d[2,1]
```

```python
# 2D array slicing
# Shape (2,2) from top right corner
arr_2d[:2,1:]
```

```python
# Shape bottom row
arr_2d[2]
```

```python
# Shape bottom row
arr_2d[2,:]
```
