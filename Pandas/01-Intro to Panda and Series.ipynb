###Â **Series**
**Important to note:**
- Series, unlike NumPy, can have axis labels, it can therefore be indexed by a label rather than just a number location.
- It can hold **any** arbitrary Python Object.

```python
import numpy as np
import pandas as pd
```

### **Creating a Series**
You can convert a list, numpy array, or a dictionary to a Series:

```python
labels = ['a','b','c']
my_data = [10,20,30]
arr = np.array(my_data)
d = {'a':10,'b':20,'c':30}
```

**Using lists**

```python
pd.Series(data = my_data)
```

```python
pd.Series(data=my_data,index=labels)
```

```python
pd.Series(my_data,labels)
```

**Using arrays**

```python
pd.Series(arr,labels)
```

```python
labels
```

**Using a dictionary**

```python
pd.Series(d)
```

### **Data in a Series**

```python
pd.Series(data=[sum,print,len])
```

### **Using an Index**

Pandas make use of index names _or_ numbers by allowing for fask look ups of information like a hash table or dictionary.

```python
ser1 = pd.Series([1,2,3,4],['USA','Germany','USSR','Japan'])
```

```python
ser1
```

```python
ser2 = pd.Series([1,2,5,3],['USA','Germany','Italy','Japan'])
```

```python
ser2
```

```python
ser1['USA']
```
