### **Pyplot Interface**

`pyplot` immitates a MATLAB interface of generating plots. This is a state-based interface, where the state - the figure - is preserved through various function calls.

When using `pyplot` to plot two variables issues arise. Quantites with varying magnitudes will dsiplay a disproportionate graph.

The solution is to separate the two quantities into different axis using an **object-oriented interface**

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

## Object-Oriented Method

`matplotlib` is also capable of instantiating figure objects and then calling methods or attributes from that object.

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8])


# Plot on that set of axes
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
# Create blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

#Larger Figure 
axes1.plot(x,y)
axes1.set_title('Larger Plot')

# Smaller Figure
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

## `subplots()`

The `plt.subplots` object will act as a more automatic axis manager.

To prevent graph **collisions** implement `.tight_layout()` to resolve.

It is good practice to do so!

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig,axes = plt.subplots(nrows=1,ncols=2) 

# Adjust the no. of plots within the figure

# subbplot() differs from figure() is that subplot() automatically...
# adds the axis for you according to the parameters you pass.

# plt.tight_layout() --> 'tight_layout()' amends any overlaps between graphs.

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout() # It is good practice to always put this at the end of your plot layouts.
```

### **Figure Size and DPI**

### **Plot Appearance**

Reference `color='your_colour'` to change the line **colour**.

Reference `linewidth='width_int'` or `lw='width_int` to change line **thickness**.

Reference `alpha='alpha_int'` to change line **transparency**.

Reference `linestyle='style'` or `ls='style'` to change line **appearance**.

Reference `marker='+'` to **highlight** each point.

Reference `markersize='your_size'` to **change** the size of the marker

Reerence `markerfacecolor='colour'` to change **inlay** colour.

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',lw=1,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
